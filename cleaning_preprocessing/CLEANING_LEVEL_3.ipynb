```python
def final_cleaning(df):
    return (
        df
        .assign(
            mileage = lambda df_ : (
                df_
                .mileage
                .str.strip()
                .str.replace(',','')
            ),

            price = lambda df_ : (
                df_
                .price
                .replace('Not Priced',np.nan)
            ),

            Exterior_color = lambda df_ : (
                df_
                .Exterior_color
                .str.lower()
                .apply(extracting_color)
            ),

            Interior_color = lambda df_ : (
                df_
                .Interior_color
                .str.lower()
                .apply(extracting_color)
            ),

            Accidents_or_damage = lambda df_ : (
                df_
                .Accidents_or_damage
                .str.strip()
                .str.lower()
                == 'at least 1 accident or damage reported'
            ),

            Clean_title = lambda df_ : (
                df_[['new_used','Clean_title']]
                .apply(is_clean_title,axis=1)
            ),

            one_owner_vehicle = lambda df_ : (
                df_[['new_used','one_owner_vehicle']]
                .apply(is_one_owner,axis=1)
            ),

            Personal_use_only = lambda df_ : (
                df_[['Personal_use_only','new_used']]
                .apply(is_personal_use_only,axis=1)
            ),

            Open_recall = lambda df_ : (
                df_[['Open_recall','new_used']]
                .apply(has_open_recall,axis=1)
            ),

            seller_name = lambda df_ : (
                df_
                .seller_name
                .str.lower()
                .str.strip()
            ),

            people_count_seller_rating = lambda df_ : (
                df_
                .people_count_seller_rating
                .apply(lambda x : x if x.isnumeric() else np.nan)
            ),

            gear_spec = lambda df_ : (
                df_
                .gear_spec
                .str.lower()
                .apply(gear_spec_extract)
                .str.extract(r'(\d+)')
            ),

            engine_size = lambda df_ : (
                df_
                .engine_size
                .str.lower()
                .str.replace('l','')
                .str.replace('-','')
                .str.strip()
            ),

            cylinder_config = lambda df_ : (
                df_
                .cylinder_config
                .str.strip()
                .str.replace('-','')
                .str.replace(' ','')
            ),

            Fuel_Type = lambda df_ : (
                df_
                .Fuel_Type
                .fillna(df_['Fuel_Type_e'])
            ),

            price_drop = lambda df_ : (
                df_
                .price_drop
                .fillna(0)
            ),

            level2_charging = lambda df_ : (
                df_
                .level2_charging
                .round()
            ),

            
            
        )
        .drop(columns='Fuel_Type_e')

)

data = final_cleaning(df)
```

### changing data types

```python
cat_cols = ['model_year','Brand_name','new_used','deal_type',
            'Exterior_color','Interior_color','Drivetrain',
            'Fuel_Type','Transmission','one_owner_vehicle',
            'Personal_use_only','Open_recall','gear_spec',
            'engine_size','cylinder_config','valves',
            'battery_range_score_cleaned','battery_range_score_category',
            'state'
            ]

numeric_cols = ['mileage','price','price_drop','MPG',
                'level2_charging','dc_fast_charging',
                'battery_capacity','excepted_range',
                'seller_rating','people_count_seller_rating',
                'car_rating','people_count_car_rating',
                'mpge_city','mpge_hwy','postal_code'
                ]

bool_cols = ['Accidents_or_damage','Clean_title']

bool_cols_sp = ['adaptive_cruise_control','power_folding_mirror',
             'cooled_seats','heated_seats','heated_steering_wheels',
             'keyless_entry','keyless_start','navigation_system',
             'remote_start','power_liftgate','automatic_parking',
             'wheel_chair_accessible','autopilot','android_auto',
             'apple_carplay','bluetooth','premium_sound_system','usb_port',
             'wifi_hotspot','satellite_radio','cd_player','homelink',
             'rear_seat_entertainment', 'dvd_player', 'alloy_wheels',
             'sunroof_moonroof', 'tow_hitch', 'tow_hooks', 'roof_rack',
             'automatic_emergency_braking', 'backup_camera', 'blind_spot_monitor',
             'brake_assist', 'led_headlights', 'lane_departure_warning',
             'rear_cross_traffic_alert', 'stability_control', 'rain_sensing_wipers',
             'leather_seats', 'memory_seat', 'third_row_seating'
             ]
final_df = (
    data
    .assign(
        **{col : data[col].astype('category') for col in cat_cols},
        **{col : pd.to_numeric(data[col]) for col in numeric_cols},
        **{col : data[col].astype(bool) for col in bool_cols},
        **{col : data[col].astype(bool).astype('boolean').replace(False,np.nan) for col in bool_cols_sp}
    )
)
```
